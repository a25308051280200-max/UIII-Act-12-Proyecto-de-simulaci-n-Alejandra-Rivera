```python
# ============================================
# PROGRAMA POO - CLASE ENTRENADOR
# Simulación sencilla de inteligencia artificial
# ============================================

class Entrenador:

    lista_entrenadores = []

    def __init__(self, nombre, deporte, equipo,
                 titulos, estrategia,
                 certificaciones, anios):

        self.nombre = nombre
        self.deporte = deporte
        self.equipo = equipo
        self.titulos = titulos
        self.estrategia = estrategia
        self.certificaciones = certificaciones
        self.anios = anios

    # ----------------------------------------
    # AGREGAR ENTRENADOR
    # ----------------------------------------
    @classmethod
    def agregar_entrenador(cls):

        print("\n=== AGREGAR ENTRENADOR ===")

        nombre = input("Nombre: ")
        deporte = input("Deporte: ")
        equipo = input("Equipo: ")
        titulos = input("Títulos ganados: ")
        estrategia = input("Estrategia: ")
        certificaciones = input("Certificaciones: ")
        anios = input("Años de carrera: ")

        nuevo = Entrenador(
            nombre,
            deporte,
            equipo,
            titulos,
            estrategia,
            certificaciones,
            anios
        )

        cls.lista_entrenadores.append(nuevo)

        print("\nEntrenador agregado correctamente.")

    # ----------------------------------------
    # LISTAR ENTRENADORES
    # ----------------------------------------
    @classmethod
    def listar_entrenadores(cls):

        print("\n=== LISTA DE ENTRENADORES ===")

        if len(cls.lista_entrenadores) == 0:
            print("No hay entrenadores registrados.")
        else:

            print("-" * 120)

            print(f"{'ID':<5} {'NOMBRE':<15} {'DEPORTE':<12} "
                  f"{'EQUIPO':<15} {'TITULOS':<10} "
                  f"{'ESTRATEGIA':<20} {'CERTIFICACIONES':<20} "
                  f"{'AÑOS':<5}")

            print("-" * 120)

            contador = 0

            for entrenador in cls.lista_entrenadores:

                print(f"{contador:<5} "
                      f"{entrenador.nombre:<15} "
                      f"{entrenador.deporte:<12} "
                      f"{entrenador.equipo:<15} "
                      f"{entrenador.titulos:<10} "
                      f"{entrenador.estrategia:<20} "
                      f"{entrenador.certificaciones:<20} "
                      f"{entrenador.anios:<5}")

                contador += 1

            print("-" * 120)

    # ----------------------------------------
    # ACTUALIZAR ENTRENADOR
    # ----------------------------------------
    @classmethod
    def actualizar_entrenador(cls):

        cls.listar_entrenadores()

        if len(cls.lista_entrenadores) > 0:

            posicion = int(input("\nID del entrenador a actualizar: "))

            if posicion < len(cls.lista_entrenadores):

                entrenador = cls.lista_entrenadores[posicion]

                print("\n=== NUEVOS DATOS ===")

                entrenador.nombre = input("Nuevo nombre: ")
                entrenador.deporte = input("Nuevo deporte: ")
                entrenador.equipo = input("Nuevo equipo: ")
                entrenador.titulos = input("Nuevos títulos: ")
                entrenador.estrategia = input("Nueva estrategia: ")
                entrenador.certificaciones = input("Nuevas certificaciones: ")
                entrenador.anios = input("Nuevos años de carrera: ")

                print("\nEntrenador actualizado correctamente.")

            else:
                print("ID no válido.")

    # ----------------------------------------
    # BORRAR ENTRENADOR
    # ----------------------------------------
    @classmethod
    def borrar_entrenador(cls):

        cls.listar_entrenadores()

        if len(cls.lista_entrenadores) > 0:

            posicion = int(input("\nID del entrenador a borrar: "))

            if posicion < len(cls.lista_entrenadores):

                eliminado = cls.lista_entrenadores.pop(posicion)

                print(f"\nEntrenador {eliminado.nombre} eliminado.")

            else:
                print("ID no válido.")

    # ----------------------------------------
    # IA SIMPLE
    # ----------------------------------------
    @classmethod
    def inteligencia_artificial(cls):

        print("\n=== IA DEL ENTRENADOR ===")

        deporte = input("¿Qué deporte practicas?: ")

        if deporte.lower() == "futbol":
            print("IA: Usa una estrategia ofensiva y entrenamiento físico.")

        elif deporte.lower() == "basquetbol":
            print("IA: Mejora la velocidad y los tiros de larga distancia.")

        elif deporte.lower() == "voleibol":
            print("IA: Practica bloqueos y trabajo en equipo.")

        elif deporte.lower() == "beisbol":
            print("IA: Entrena reflejos y precisión en bateo.")

        else:
            print("IA: Mantén disciplina y entrenamiento constante.")

# ==================================================
# MENÚ PRINCIPAL
# ==================================================

opcion = ""

while opcion != "6":

    print("\n")
    print("====================================")
    print(" SISTEMA INTELIGENTE DE ENTRENADORES ")
    print("====================================")
    print("1. Agregar entrenador")
    print("2. Listar entrenadores")
    print("3. Actualizar entrenador")
    print("4. Borrar entrenador")
    print("5. IA deportiva")
    print("6. Salir")
    print("====================================")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        Entrenador.agregar_entrenador()

    elif opcion == "2":
        Entrenador.listar_entrenadores()

    elif opcion == "3":
        Entrenador.actualizar_entrenador()

    elif opcion == "4":
        Entrenador.borrar_entrenador()

    elif opcion == "5":
        Entrenador.inteligencia_artificial()

    elif opcion == "6":
        print("\nPrograma finalizado.")

    else:
        print("\nOpción incorrecta.")
```
